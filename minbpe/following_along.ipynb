```python
# we can't use the raw bytes of utf-8 as we want to support a larger vocabulary size. How?
# use the byte pair encoding algorithm
text = "Manta rays are large rays belonging to the genus Mobula (formerly its own genus Manta). Three species are known: M. birostris, the largest at 7 m (23 ft) in width, M. yarae, which reaches 6 m (20 ft), and M. alfredi, the smallest at 5.5 m (18 ft). All three have triangular pectoral fins, horn-shaped cephalic fins and large, forward-facing mouths. They are classified among the Myliobatiformes (stingrays and relatives) and are placed in the family Myliobatidae (eagle rays). They have one of the highest brain-to-body ratio of all fish, and can pass the mirror test. "
tokens = text.encode("utf-8")
tokens = list(map(int, tokens))
# print(f"{text=}")
# print(f"{len(text)=}")
# print(f"{tokens=}")
# print(f"{len(tokens)=}")


```

```python
# write a function that given a list of tokens finds the most common pair
# Andrej calls this "get_stats"
from collections import Counter
def get_most_common_pair(tokens):
    num_pairs = Counter()
    for i, token in enumerate(tokens[:-1]):
        num_pairs[(token, tokens[i+1])] += 1
    # return num_pairs # return whole dict
    return num_pairs.most_common(1)[0][0]

print(f"{get_most_common_pair(tokens)=}")
```

```python
# replace all instances of most common pair with newly minted token
def merge(ids, pair_to_merge, merged_pair):
    merged_ids = []
    i=0
    while i < len(ids):
        if i < len(ids) - 1 and ids[i] == pair_to_merge[0] and ids[i+1] == pair_to_merge[1]:
            merged_ids.append(merged_pair)
            i += 2
        else:
            merged_ids.append(ids[i])
            i += 1
    return merged_ids

print(merge([5, 6, 6, 7, 9, 1], (6,7), 99))
```
